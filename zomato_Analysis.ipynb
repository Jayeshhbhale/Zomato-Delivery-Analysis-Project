```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv(r"zomato dataset.csv")
```

```python
df.head()
```

```python
# making values to numeric 
df['Delivery_person_Age'] = pd.to_numeric(df['Delivery_person_Age'], errors='coerce')
df['Time_taken (min)'] = pd.to_numeric(df['Time_taken (min)'], errors='coerce')
```

### Plotting Delivery Time Distribution

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Time_taken (min)'], bins=20, kde=True)
plt.title('Distribution of Delivery Time')
plt.xlabel('Time taken (min)')
plt.ylabel('Frequency')
plt.show()
```

### Scatter Plot: Age vs Delivery Time

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Delivery_person_Age', y='Time_taken (min)', data=df)
plt.title('Delivery Person Age vs Delivery Time')
plt.xlabel('Delivery Person Age')
plt.ylabel('Time taken (min)')
plt.show()
```

### Box Plot: Delivery Time by Weather Conditions

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Weather_conditions", y="Time_taken (min)",data=df)
plt.title('Delivery Time by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Time Taken (Min)')
plt.xticks(rotation=45)
plt.show()
```

### Road Traffic Density vs Delivery Time

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Road_traffic_density", y="Time_taken (min)",data=df)
plt.title('Avg Delivery Time by Road Traffic Density')
plt.xlabel('Road Traffic Density')
plt.ylabel('Time Taken (Min)')
plt.show()
```

### Type of Vehicle and Order Distribution

```python
# Type of vehicle distribution
plt.figure(figsize=(8,5))
sns.countplot(x="Type_of_vehicle", data=df)
plt.title('Type of vehicle distribution')
plt.xlabel("Vehicle Type")
plt.ylabel("Count")
plt.show()
```

```python
# Type of Order distribution
plt.figure(figsize=(8,5))
sns.countplot(x="Type_of_order", data=df)
plt.title('Type of Order distribution')
plt.xlabel("Order Type")
plt.ylabel("Count")
plt.show()
```

### Delivery Time on Festival Days

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Festival', y='Time_taken (min)', data=df)
plt.title('Delivery Time by Festival Days')
plt.xlabel('Festival')
plt.ylabel('Time taken (min)')
plt.show()
```

### Vehicle Condition vs Delivery Time

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Vehicle_condition', y='Time_taken (min)', data=df)
plt.title('Delivery Time by Vehicle Condition')
plt.xlabel('Vehicle Condition')
plt.ylabel('Time Taken (min)')
plt.show()
```
